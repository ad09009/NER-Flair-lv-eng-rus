```python
from flair.models import SequenceTagger
from flair.data import Sentence
from flair.datasets import CONLL_03
from flair.data import Corpus
from flair.datasets import ColumnCorpus
```

```python
# Paths

# all folder paths need to be extended with drive location when running on google drive
#drive_path = 'drive/MyDrive/'

lv_data_folder = 'resources/taggers/lv-ner-flair-edit'
rus_data_folder = 'resources/taggers/rus-ner-flair-edit'
eng_data_folder = 'resources/taggers/eng-ner-flair-edit'

eng_corpus_base_path = 'resources/tasks'

lv_model_path = lv_data_folder + '/final-model.pt'
eng_model_path = eng_data_folder + '/best-model.pt'
rus_model_path = rus_data_folder + '/final-model.pt'
```

```python
# Load the data for testing
```

```python
# ENG corpus

corpus: Corpus = CONLL_03(base_path=eng_corpus_base_path)

print(corpus)
```

```python
# RUS corpus

# define columns
columns_rus = {0: 'text', 1: 'pos', 2: 'pos2' , 3: 'ner'}

# Data from https://competitions.codalab.org/competitions/36044#participate-get_data
corpusrus: Corpus = ColumnCorpus(rus_data_folder, columns_rus,
                              train_file='train.txt',
                              test_file='test.txt',
                              dev_file='dev.txt')
print(corpusrus)
```

```python
# LV corpus

# define columns
columns_lv = {0: 'text', 1: 'ner'}

corpuslv: Corpus = ColumnCorpus(lv_data_folder, columns_lv,
                              train_file='train.txt',
                              test_file='test.txt',
                              dev_file='dev.txt')
print(corpuslv)
```

```python
# Load the trained models

bestmodellv = SequenceTagger.load(lv_model_path)
bestmodelrus = SequenceTagger.load(rus_model_path)
bestmodel = SequenceTagger.load(eng_model_path)
```

```python
# Print model details
```

```python
# ENG

bestmodel.eval()
```

```python
# LV

bestmodellv.eval()
```

```python
# RUS

bestmodelrus.eval()
```

```python
# Run evaluations
```

```python
# ENG evaluation
result_eng = bestmodel.evaluate(corpus.test, mini_batch_size=32, gold_label_type='ner', out_path=f"predictions_eng.txt")
print(result_eng.detailed_results)
```

```python
# RUS evaluation
result_rus = bestmodelrus.evaluate(corpusrus.test, mini_batch_size=32, gold_label_type='ner', out_path=f"predictions_rus.txt")
print(result_rus.detailed_results)
```

```python
# LV evaluation
result_lv = bestmodellv.evaluate(corpuslv.test, mini_batch_size=32, gold_label_type='ner', out_path=f"predictions_lv.txt")
print(result_lv.detailed_results)
```

```python
# RUS test sentences
sentence_listrus = []
sentence_listrus.append(Sentence("              ,       ."))
sentence_listrus.append(Sentence("                ,        "))
sentence_listrus.append(Sentence("   (),     ,           ,      ."))
sentence_listrus.append(Sentence("      :          (1984),     ."))
```

```python
# RUS demo
for sentrus in sentence_listrus:
    bestmodelrus.predict(sentrus)
    print(sentrus)
    for entity in sentrus.get_spans():
        print(entity)
    print('\n')
```
